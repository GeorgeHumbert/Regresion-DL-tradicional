```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Cargar el modelo entrenado
model = tf.keras.models.load_model('airline_passengers_lstm.h5')

# Cargar los datos de prueba (suponiendo que ya los hayas preprocesado)
X_test = np.load('data/X_sequences.npy')
y_test = np.load('data/y_sequences.npy')

# Realizar predicciones
y_pred = model.predict(X_test)

# Invertir la normalización si fue aplicada durante el preprocesamiento (ajustar según sea necesario)
# y_pred = scaler.inverse_transform(y_pred)
# y_test = scaler.inverse_transform(y_test)

# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Valores Reales')
plt.plot(y_pred, label='Predicciones', linestyle='--')
plt.title('Predicciones vs Valores Reales')
plt.xlabel('Tiempo')
plt.ylabel('Número de Pasajeros')
plt.legend()
plt.show()
```
